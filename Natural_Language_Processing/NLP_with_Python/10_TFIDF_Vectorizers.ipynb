## Vectorizing Raw Data: Tfidf Vectorization

#### TFIDF vectorization

```python
import pandas as pd
import re
import string
import nltk

stopword = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()

pd.set_option('display.max_colwidth', 100)
```

```python
data = pd.read_csv('data/SMSSpamCollection.tsv', sep='\t', header=None)
data.columns = ['label', 'body_text']

data.head()
```

```python
def clean_text(text):
    text = "".join([char for char in text if char not in string.punctuation])
    tokenize = re.split('\W+', text)
    text = [ps.stem(word) for word in tokenize if word not in stopword]
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(analyzer=clean_text)

X_tfidf = tfidf_vect.fit_transform(data['body_text'])

print(X_tfidf.shape)
print(tfidf_vect.get_feature_names())
```

### Apply TfiIdf Vectorizer to smaller sample

```python
data_sample = data[:20]

tfidf_vect_sample = TfidfVectorizer(analyzer=clean_text)

X_tfidf_sample = tfidf_vect_sample.fit_transform(data_sample['body_text'])

print(X_tfidf_sample.shape)
print(tfidf_vect_sample.get_feature_names())
```

### Vectorizers output sparse matrices

**Sparse Matrix:** A matrix in which most entries are 0. In the interest of efficient storage, a sparse matrix will be stored by only storing the locations of the non-zero elements.

```python
X_tfidf_df = pd.DataFrame(X_tfidf_sample.toarray(), columns=tfidf_vect_sample.get_feature_names())

X_tfidf_df.head(10)
```
