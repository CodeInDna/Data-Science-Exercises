## Building Machine Learning Classifiers: Building a basic Random Forest Model

### Read in & clean text

```python
import pandas as pd
import re
import string
from nltk.corpus import stopwords
from nltk import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

import matplotlib.pyplot as plt

pd.set_option('display.max_colwidth', 100)
%matplotlib inline
```

```python
data = pd.read_csv('data/SMSSpamCollection.tsv', sep='\t', header=None)
data.columns = ['label', 'body_text']

data.head()
```

```python
stopword = stopwords.words('english')
ps = PorterStemmer()

def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation])
    
    return round(count/(len(text) - text.count(" ")), 3) * 100

def clean_text(text):
    text = "".join([char for char in text if char not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopword]
    
    return text

data['punct%'] = data['body_text'].apply(lambda x: count_punct(x))
data['body_len'] = data['body_text'].apply(lambda x: len(x) - x.count(" "))

data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(analyzer=clean_text)
X_counts = count_vect.fit_transform(data['body_text'])

print(X_counts.shape)
print(count_vect.get_feature_names())
```

```python
tfidf = TfidfVectorizer(analyzer=clean_text)
X_tfidf = tfidf.fit_transform(data['body_text'])

print(X_tfidf.shape)
print(tfidf.get_feature_names())
```

```python
X_temp = pd.DataFrame(X_tfidf.toarray())
X_temp.columns = tfidf.get_feature_names()

X_temp.head()
```

```python
X_features = pd.concat([data['body_len'], data['punct%'], X_temp], axis=1)

X_features.head()
```

### Explore Random Forest Classifier Attributes and Hyperparameters

### Explore RandomForestClassifier through Cross-Validation

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
rf = RandomForestClassifier(n_jobs=-1)
kfold = KFold(n_splits=5)
cv = cross_val_score(rf, X_features, data['label'], cv=kfold, scoring='accuracy', n_jobs=-1)
cv
```

### Explore RandomForestClassifier through Holdout-Set

```python
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(X_features, data['label'], test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=50, max_depth=20, n_jobs=-1)
rfc_model = rfc.fit(train_x, train_y)
```

```python
sorted(zip(rfc.feature_importances_, train_x.columns), reverse=True)[:10]
```

```python
pred_y = rfc_model.predict(test_x)
precision, recall, fscore, support = score(test_y, pred_y, pos_label='spam', average='binary')
```

```python
print(f"Precision: {round(precision, 3)}, Recall: {round(recall, 3)}, Accuracy: {round(sum(pred_y==test_y)/len(pred_y), 3)}")
```

### Explore RandomForestClassifier with GridSearCV

```python
def train_rf(n_est, depth):
    rfc = RandomForestClassifier(n_estimators=n_est, max_depth=depth, n_jobs=-1)
    rfc_model = rfc.fit(train_x, train_y)
    pred_y = rfc_model.predict(test_x)
    precision, recall, fscore, support = score(test_y, pred_y, pos_label='spam', average='binary')
    print(f"Est: {n_est} / Depth: {depth}-----Precision: {round(precision, 3)}, Recall: {round(recall, 3)}, Accuracy: {round(sum(pred_y==test_y)/len(pred_y), 3)}")
```

```python
for n_est in [10, 50, 100]:
    for depth in [10, 20, 30, None]:
        train_rf(n_est, depth)
```

```python
from sklearn.model_selection import GridSearchCV

rfc = RandomForestClassifier()
params = {'n_estimators': [10, 150, 300], 'max_depth': [30, 60, 90, None]}

gs = GridSearchCV(rfc, params, cv=5, n_jobs=-1)
gs.fit(X_tfidf, data['label'])
pd.DataFrame(gs.cv_results_).sort_values('mean_test_score', ascending=False)[:5]
```
