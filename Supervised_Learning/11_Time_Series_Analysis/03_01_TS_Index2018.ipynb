The columns contains the closing price of 4 market indexes. 
Each market indexes is a portfolio of most traded public companies on the respective stock exchange.

S&P 500 - spx (USA)

DAX 30 - dax (Germany)

FTSE 100 - ftse (UK)

NIKKEI 225 - nikkei (Japan)

Each market indexes contains the number of companies in the portfolio of the respective stock exchange. for e.g. S&P contains 500 companies and so on.


### Import the Packages

```python
import pandas as pd
import numpy as np

import scipy.stats
import pylab

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Importing the Data

```python
raw_csv_data = pd.read_csv('data/Index2018.csv')
```

```python
df_comp = raw_csv_data.copy()
```

### Plotting the Data

```python
df_comp.spx.plot(figsize=(20, 5), title="S&P 500 Prices")
plt.show()
```

```python
df_comp.ftse.plot(figsize=(20, 5), title="FTSE 100 Prices")
plt.show()
```

```python
df_comp.spx.plot(figsize=(20, 5), title="S&P 500 Prices")
df_comp.ftse.plot(figsize=(20, 5), title="FTSE 100 Prices")
plt.title('S&P VS FTSE')
plt.show()
```

### The QQ Plot

```python
scipy.stats.probplot(df_comp.spx, plot=pylab)
pylab.show()
```
