## Bivariate Scatter Plot and Correlation

```python
# import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# load data
fuel_econ = pd.read_csv('../../Datasets/fuel_econ.csv')

fuel_econ.head()
```

**Using Matplotlib**

```python
plt.scatter(data=fuel_econ, x='displ', y='comb')
plt.xlabel('Displacement')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```

**Using Seaborn**

```python
fig = plt.figure(figsize=[10, 4])

plt.subplot(1, 2, 1)
sb.regplot(data=fuel_econ, x='displ', y='comb')
plt.title('With Regression Line')
plt.xlabel('Displacement')
plt.ylabel('Combined Fuel Efficiency (mpg)');

plt.subplot(1, 2, 2)
sb.regplot(data=fuel_econ, x='displ', y='comb', fit_reg=False)
plt.title('Without Regression Line')
plt.xlabel('Displacement')
plt.ylabel('Combined Fuel Efficiency (mpg)');
```

### Overplotting, Transparency and Jitter

**Using Matplotlib for transparency**

**NOTE:** There is no jitter behaviour in Matplotlib Module

```python
plt.scatter(data=fuel_econ, x='year', y='comb', alpha=1/20);
```

**Using Seaborn for Transparency and Jitter**

```python
sb.regplot(data=fuel_econ, x='year', y='comb', scatter_kws={'alpha':1/20}, x_jitter=0.3, y_jitter=0.2);
```
