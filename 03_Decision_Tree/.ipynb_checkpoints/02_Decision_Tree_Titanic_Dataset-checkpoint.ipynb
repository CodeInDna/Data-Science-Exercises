## Titanic Survival Exploration with Decision Trees

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
import random

# Pretty Display for Notebooks
%matplotlib inline

# Set a Random Seed
random.seed(42)

# Load the Dataset
address = '../Datasets/decision_trees_titanic_data.csv'
titanic = pd.read_csv(address)

# Print the First Few Entries of the Titanic Data
titanic.head()
```

Recall that these are the various features present for each passenger on the ship:

* **Survived:** Outcome of survival (0 = No; 1 = Yes)
* **Pclass:**  Socio-economic class (1 = Upper class; 2 = Middle class; 3 = Lower class)
* **Name:**  Name of passenger
* **Sex:**  Sex of the passenger
* **Age:**  Age of the passenger (Some entries contain NaN)
* **SibSp:**  Number of siblings and spouses of the passenger aboard
* **Parch:**  Number of parents and children of the passenger aboard
* **Ticket:**  Ticket number of the passenger
* **Fare:**  Fare paid by the passenger
* **Cabin:** Cabin number of the passenger (Some entries contain NaN)
* **Embarked:**  Port of embarkation of the passenger (C = Cherbourg; Q = Queenstown; S = Southampton)

Since we're interested in the outcome of survival for each passenger or crew member, we can remove the Survived feature from this dataset and store it as its own separate variable outcomes. We will use these outcomes as our prediction targets. Run the code cell below to remove Survived as a feature of the dataset and store it in outcomes.

```python
# Store the 'Survived' Feature in a new variable and remove it from the dataset
outcomes = titanic.Survived
features_raw = titanic.drop('Survived', axis=1)

# Show the new dataset with 'Survived' removed
features_raw.head()
```

**Preprocessing the Data**

Now, let's do some data preprocessing. First, we'll one hot encode the features

```python
features = pd.get_dummies(features_raw)

features.head()
```

And now we'll fill in any blanks with zeroes

```python
features = features.fillna(0)

features.head()
```

```python
# Split the data 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, outcomes, test_size=0.2, random_state=42)
```
