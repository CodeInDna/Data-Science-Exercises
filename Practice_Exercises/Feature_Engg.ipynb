```python
dfA = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

dfB = pd.DataFrame({'key': ['K4', 'K1', 'K2', 'K3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
dfA
```

```python
dfB
```

```python
pd.merge(dfA, dfB, on='key') # inner join
```

```python
pd.merge(dfA, dfB, how='right', on='key') # right join
```

```python
pd.merge(dfA, dfB, how='outer', on='key')
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                      'B': ['B0', 'B1', 'B2']},
                   index = ['K0', 'K1', 'K2'])

left
```

```python
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                       'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])

right
```

```python
left.join(right)
```

```python
right.join(left)
```
